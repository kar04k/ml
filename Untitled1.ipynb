```python
import os
import requests
import pandas as pd
from tqdm import tqdm

# Load metadata CSV
metadata_df = pd.read_csv('/content/train.csv')

# Filter for the first 2000 entries
metadata_df_2000 = metadata_df.head(2000)

# Create a directory for downloaded images
save_folder = '/content/images1'
os.makedirs(save_folder, exist_ok=True)

def download_image(image_url, save_path):
    try:
        response = requests.get(image_url)
        response.raise_for_status()  # Check for HTTP errors
        with open(save_path, 'wb') as file:
            file.write(response.content)
    except Exception as e:
        print(f"Error downloading {image_url}: {e}")

# Download and save images
for index, row in tqdm(metadata_df_2000.iterrows(), total=len(metadata_df_2000)):
    image_url = row['image_link']
    group_id = row['group_id']
    image_name = f"{group_id}_{index}.jpg"  # Name image with group ID and index
    save_path = os.path.join(save_folder, image_name)
    download_image(image_url, save_path)
```

```python
import cv2
import os
import numpy as np

# Function to preprocess a single image
def preprocess_image(image_path, target_size=(224, 224)):
    # Read the image
    image = cv2.imread(image_path)

    # Resize the image
    image_resized = cv2.resize(image, target_size)

    # Normalize the image (scale pixel values between 0 and 1)
    image_normalized = image_resized / 255.0

    return image_normalized

# Function to preprocess all images in a folder
def preprocess_images_from_folder(image_folder, save_folder, target_size=(224, 224)):
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    for image_name in os.listdir(image_folder):
        image_path = os.path.join(image_folder, image_name)
        if os.path.isfile(image_path):
            processed_image = preprocess_image(image_path, target_size)

            # Save the preprocessed image (optional)
            save_path = os.path.join(save_folder, image_name)
            cv2.imwrite(save_path, processed_image * 255)  # Convert back to [0-255] range for saving
            print(f"Processed and saved {image_name}")

# Define paths
image_folder = '/content/images1'  # Folder with downloaded images
preprocessed_folder = '/content/images'  # Folder to save preprocessed images

# Preprocess the images
preprocess_images_from_folder(image_folder, preprocessed_folder, target_size=(224, 224))
```

```python
import pandas as pd
import re

# Load your dataset
# Assuming 'train.csv' is the file containing your training data
train_df = pd.read_csv('/content/train.csv')

# Helper function to split entity_value into numeric value and unit
def split_value_unit(entity_value):
    match = re.match(r"([0-9\.]+)\s*(.*)", entity_value)
    if match:
        return float(match.group(1)), match.group(2)
    else:
        return None, None

# Split the entity_value into numeric and unit for training
train_df['value'], train_df['unit'] = zip(*train_df['entity_value'].apply(split_value_unit))

# Handle missing values by dropping rows with NaN values in 'value' or 'unit'
train_df = train_df.dropna(subset=['value', 'unit'])

# Save the processed dataframe to a new CSV file
train_df.to_csv('/content/metadata.csv', index=False)

print("Preprocessing complete. Processed data saved to 'processed_train.csv'.")
```

```python
import os
import pandas as pd

# Define file paths
metadata_file = '/content/metadata.csv'
image_folder = '/content/images'

# Load metadata CSV
metadata_df = pd.read_csv(metadata_file)

# List all image files in the image directory
image_files = os.listdir(image_folder)

# Convert list of image files to a set for faster lookup
image_files_set = set(image_files)

# Create a column with expected image file names
metadata_df['expected_file_name'] = metadata_df['group_id'].astype(str) + '_' + metadata_df.index.astype(str) + '.jpg'

# Check if each image in metadata exists in the image folder
metadata_df['file_exists'] = metadata_df['expected_file_name'].isin(image_files_set)

# Filter out rows where the image does not exist
filtered_metadata = metadata_df[metadata_df['file_exists']]

# Save the filtered metadata
filtered_metadata.to_csv('filtered_metadata.csv', index=False)

print(f"Total metadata entries: {len(metadata_df)}")
print(f"Filtered metadata entries (with images): {len(filtered_metadata)}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the filtered metadata CSV
metadata_df = pd.read_csv('filtered_metadata.csv')

# Define the proportions for splitting
train_size = 0.8
val_size = 0.1
test_size = 0.1

# First split: training and temporary (validation + test)
train_df, temp_df = train_test_split(metadata_df, test_size=(val_size + test_size), random_state=42)

# Second split: validation and test from the temporary dataset
val_df, test_df = train_test_split(temp_df, test_size=test_size/(val_size + test_size), random_state=42)

# Save the splits to CSV files
train_df.to_csv('train_metadata.csv', index=False)
val_df.to_csv('val_metadata.csv', index=False)
test_df.to_csv('test_metadata.csv', index=False)

# Print the size of each split
print(f"Training set size: {len(train_df)}")
print(f"Validation set size: {len(val_df)}")
print(f"Test set size: {len(test_df)}")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define paths to the directories
train_dir = 'path/to/train/images'
val_dir = 'path/to/val/images'
test_dir = 'path/to/test/images'

# Create ImageDataGenerator instances
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Create data generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'  # Use 'categorical' for multi-class classification
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```
